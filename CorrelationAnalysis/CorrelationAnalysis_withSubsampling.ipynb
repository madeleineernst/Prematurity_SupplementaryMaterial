## Univariate correlation at the individual metabolite level using Kendall’s Tau including subsampling to equalize statistical power

load blank filtered and batch normalised feature table

```r
ft <- read.csv('../../featuretable_blankfiltered_batchnormalised.tsv', sep = '\t', check.names = F)
```

```r
dim(ft)
```

load metadata

```r
md <- read.table('../../metadata.tsv', sep = '\t', comment.char = '', check.names = F, header = T)
```

```r
dim(md)
```

are there any features with all zeros?

```r
length(which(rowSums(ft[,2:ncol(ft)]) == 0))
```

transpose feature table for correlation analysis

```r
ft <- as.data.frame(t(ft))
colnames(ft) <- ft[1,]
ft <- ft[-1,]
```

match metadata with feature table and bring filenames in same order (this should return TRUE)

```r
md <- md[which(md$'#SampleID' %in% rownames(ft)),]
md <- md[match(rownames(ft),md$'#SampleID'),]
identical(as.character(md$'#SampleID'),as.character(rownames(ft)))
```

how many samples are there per week of gestational age?

```r
table(md$Gestationsalder)
```

randomly subsample 10 samples per week of gestational age to equalize statistical power

```r
nperm = 1000
```

```r
pvalscor <- matrix('none',ncol(ft),nperm)
rownames(pvalscor) <- colnames(ft)
```

```r
estcor <- matrix('none',ncol(ft),nperm)
rownames(estcor) <- colnames(ft)
```

```r
for (i in 1:nperm){
    
    md_subs <- md[-which(md$Gestationsalder == 30),]
    md_subs <- md_subs %>% group_by(Gestationsalder) %>% sample_n(10)
    md_subs <- bind_rows(md[which(md$Gestationsalder == 30),],md_subs)
    
    ftt <- ft[which(rownames(ft) %in% md_subs$'#SampleID'),]
    md_subs <- md_subs[match(rownames(ftt),md_subs$'#SampleID'),]
    
    if (length(which(colSums(ftt)==0)) != 0){
        ftt <- ftt[,-which(colSums(ftt)==0)]
        }
    
    message(paste(i,identical(as.character(md_subs$'#SampleID'),as.character(rownames(ftt))), sep = ' '))
    message(table(md_subs$Gestationsalder))
    
    catscont <- md_subs$Gestationsalder
    
    corres <- t(sapply(1:ncol(ftt), function(x) unlist(cor.test(catscont, scale(ftt[,x])[,1], method = 'kendall')[c("estimate", "p.value")])))
    #corres <- t(sapply(1:ncol(ftt), function(x) unlist(cor.test(catscont, ftt[,x], method = 'kendall')[c("estimate", "p.value")])))

    corres <- cbind(corres,p.adjust(corres[,2], method = "fdr"))
    rownames(corres) <- colnames(ftt)
    corres <- cbind(corres,rownames(corres))
    colnames(corres) <- c('estimate.cor','p.value','p.value.corrected','cluster id')
    
    pvalscor[match(rownames(corres),rownames(pvalscor)),i] <- corres[,3]
    estcor[match(rownames(corres),rownames(estcor)),i] <- corres[,1]
    
}
```

```r
class(estcor) <- "numeric"
class(pvalscor) <- "numeric"
```

```r
pvalscor <- cbind(rowSums(pvalscor <= 0.05),pvalscor)
```

```r
pvalscor <- cbind(rowSums(pvalscor <= 0.05, na.rm = T),pvalscor)
```

```r
estcor <- cbind(rowMeans(estcor,na.rm = TRUE),apply(estcor, 1, min, na.rm = T), apply(estcor, 1, max, na.rm = T), estcor)
```

```r
colnames(pvalscor) <- c('n_below005',paste(rep('pvalcor', nperm),1:nperm, sep = '_'))
colnames(estcor) <- c('Mean', 'Min', 'Max', paste(rep('estimatedcor', nperm),1:nperm, sep = '_'))
estcor <- cbind(rownames(estcor),estcor)
colnames(estcor)[1] <- 'clusterID'
pvalscor <- cbind(rownames(pvalscor),pvalscor)
colnames(pvalscor)[1] <- 'clusterID'
colnames(pvalscor)[1] <- 'n_below005_NAremoved'
pvalscor <- cbind(apply(pvalscor[,4:ncol(pvalscor)], 1, function(x) sum(is.na(x))),pvalscor)
colnames(pvalscor)[1] <- 'numberofNAs'
```

```r
write.table(estcor, 'KendallsTauCorrelation_GA_1000Perm.txt', sep = '\t', quote = F, row.names = F)
```

```r
write.table(pvalscor, 'KendallsTauCorrelation_pval_GA_1000Perm.txt', sep = '\t', quote = F, row.names = F)
```
